```python
from tools import get_corpus, prepare_corpus, get_word2vec_model
```

```python
word2vec_model = get_word2vec_model()
```

```python
corpus = get_corpus()
corpus
```

```python
prepared_corpus = prepare_corpus(corpus)
```

```python
from sklearn.metrics import pairwise_distances

def corpus_distance_matrix(prepared_corpus):
    X = [[i] for i in range(0, len(corpus))]
    distance = lambda x, y: word2vec_model.wmdistance(prepared_corpus[int(x[0])], prepared_corpus[int(y[0])])
    return pairwise_distances(X, metric=distance)
```

```python
distance_matrix = corpus_distance_matrix(prepared_corpus)
```
