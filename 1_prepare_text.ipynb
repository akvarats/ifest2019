```python
text = "Не горит фонарь уличного освещения над пешеходным переходом в районе остановки Мясокомбинат в сторону Центра. Просьба восстановить освещение."
```

```python
from string import punctuation

punctuation
```

```python
import nltk

words = [w for w in nltk.word_tokenize(text, language="russian") if w not in punctuation]

words
```

```python
from rnnmorph.predictor import RNNMorphPredictor

predictor = RNNMorphPredictor(language="ru")

morphs = predictor.predict(words)

norm_pos = ["{}_{}".format(m.normal_form, m.pos) for m in morphs]

norm_pos
```
